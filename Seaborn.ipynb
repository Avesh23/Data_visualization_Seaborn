```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris =  sns.load_dataset("iris")
planets= sns.load_dataset("planets")
titanic= sns.load_dataset("titanic")
penguins = sns.load_dataset("penguins")
car_crashes = sns.load_dataset("car_crashes")
```

```python
tips
```

LINE PLOT

```python
sns.lineplot(x="day",y="tip",data=tips)
```

BAR PLOT


```python
sns.barplot(x="day" ,y="total_bill", hue="sex",data=tips )
```

```python
sns.barplot(x="day" ,y="total_bill", data=tips ,hue="sex",errorbar=("ci",0))
#plt.legend(title='sex', loc='upper right', bbox_to_anchor=(1, 1))
plt.legend(title='Sex', loc='upper left', bbox_to_anchor=(1, 1))
```

HISTOGRAM PLOT (HIST)


```python
penguins
```

```python
sns.histplot(data=penguins, x="bill_length_mm",y="bill_depth_mm",bins=30)
plt.xlabel("Bill length")
plt.ylabel("Bill Depth")
```

```python
sns.histplot(data=tips, x="day",hue="sex",palette="GnBu")
plt.xlabel("sex")
plt.ylabel("count")
```

SCATTER PLOT


```python
sns.scatterplot(data=tips, x="total_bill",y="tip", hue="time",size="size",palette='GnBu')
```
